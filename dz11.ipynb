```python
from sympy import*
```

```python
x = Symbol("x")
```

## 1. Определить корни

```python
f = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
f
```

```python
solveset(f)
```

## 2. Найти интервалы, на которых функция возрастает и убывает

```python
dif_func = trigsimp(diff(f))
eq = Eq(dif_func, 0)
eq
```

```python
for i in range (len(res_dif)-1):
    if dif_func.subs(x, (res_dif[i] + res_dif[i+1])/2) > 0:
        print(f"Функция возрастает от {res_dif[i]} до {res_dif[i+1]}")
    else:
        print(f"Функция убывает от {res_dif[i]} до {res_dif[i+1]}")
```

## 3. Построить график

```python
plot(f,dif_func, (x,-pi,2*pi))
```

## 4. Вычислить вершину

```python
z = sorted(solveset(eq, x, Interval(-pi, pi)))
res_dif = []
for elem in z:
    res_dif.append(simplify(elem))
pretty_print(res_dif)    
```
